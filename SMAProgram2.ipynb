```python
import networkx as nx
import matplotlib.pyplot as plt

def link_prediction_jaccard(graph):
    # Calculate Jaccard coefficient for all pairs of nodes
    jaccard_coefficients = list(nx.jaccard_coefficient(graph))

    # Sort the coefficients in descending order
    sorted_coefficients = sorted(jaccard_coefficients, key=lambda x: x[2], reverse=True)

    return sorted_coefficients

# Create an example graph
graph = nx.Graph()
graph.add_edges_from([(1, 2), (1, 3), (1, 4), (2, 4), (2, 5), (3, 4), (4, 5), (4, 6)])

# Perform link prediction using Jaccard coefficient
predicted_links = link_prediction_jaccard(graph)

# Display the top predicted links
print("Top Predicted Links:")
for u, v, score in predicted_links[:10]:
    print(f"Node {u} and Node {v}, Jaccard Coefficient: {score:.4f}")

# Draw and display the graph
plt.figure(figsize=(8, 6))
nx.draw(graph, with_labels=True, node_color='lightblue', edge_color='gray', font_weight='bold')
plt.title("Graph Visualization")
plt.show()
```
